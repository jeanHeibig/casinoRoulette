```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def r0mu0(n):
    P = np.append(n, -np.ones(n - 1))
    r = np.roots(P)
    W = np.vander(r ** -1)
    mu = -np.ones(n - 1) @ np.linalg.inv(W)
    return np.real(np.array([r[0], mu[0]]))
```

```python
M = 90
N = np.arange(2, M)
v_r0mu_0 = np.zeros((M - 2, 2))
for n in N:
    v_r0mu_0[n - 2] = r0mu0(n)
```

```python
r0, mu0 = v_r0mu_0[:, 0], v_r0mu_0[:, 1]
```

```python
plt.scatter(N, np.log(1 - r0))
```

```python
plt.scatter(N, 1 + mu0)
```

```python
k_sim = lambda q: np.log((q - 1) / (mu0)) / np.log(r0)
```
