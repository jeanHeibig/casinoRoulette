Notons $(u_{m})_{m \in \mathbb{N}}$ la fortune du casino en jetons à l'instant $m$, avec $u_{0} > 0$.

L'évolution de $u$ est décrite par la relation de récurrence suivante :

$$u_{m + 1} = u_{m} + T_{m}$$

$(T_{m})_{m \in \mathbb{N}}$ désigne une suite de variables aléatoires indépendantes et distribuées selon la loi suivante :

$$T_{m} = \begin{cases}
  -d \textrm{ avec probabilité } f_{d} p_{d}, d \in [\![1, n - 1]\!]
\\1 \textrm{ avec probabilité } p_{g}
\end{cases}$$

Les $(f_{d})_{d \in [\![1, n - 1]\!]}$ désignent les fréquences de mise des joueurs, et $\sum_{d = 1}^{n - 1} f_{d} = 1$. À la roulette française, les valeurs de mise autorisées sont $d \in \lbrace 1/2, 1, 2, 5, 8, 11, 17, 35 \rbrace$.

Les $(p_{d})_{d \in [\![1, n - 1]\!]}$ désignent les probabilités de gain de la somme $d$ par le joueur, avec à la roulette $\forall d \in [\![1, n - 1]\!], p_{d} = \frac{n}{(d + 1)(n + z)}$ où $z \in \mathbb{N}^{*}$ désigne le nombre de zéros sur la roulette. À la roulette française, $z = 1$.

Inversement, $p_{g}$ désigne la probabilité de gain du casino, vérifiant $p_{g} + \sum_{d = 1}^{n - 1} f_{d} p_{d} = 1$.

Vérifions que l'espérance du casino est positive :

$$\begin{align*}\mathbb{E}(T_{m}) =
  &=p_{g} - \sum_{d = 1}^{n - 1} d f_{d} p_{d}
\\&=1 - \sum_{d = 1}^{n - 1} f_{d} p_{d} - \sum_{d = 1}^{n - 1} d f_{d} p_{d}
\\&=1 - \sum_{d = 1}^{n - 1} (d + 1) f_{d} p_{d}
\\&=1 - \sum_{d = 1}^{n - 1} (d + 1) f_{d} \frac{n}{(d + 1)(n + z)}
\\&=1 - \frac{n}{(n + z)} \sum_{d = 1}^{n - 1} f_{d}
\\&=\frac{z}{(n + z)}
\\&>0
\end{align*}$$

Nous disons que le casino survit s'il dispose d'une fortune strictement positive à chaque instant, c'est-à-dire $\forall m \in \mathbb{N}, u_{m} > 0$.

Notons $q_{k}$ la probabilité de survie du casino, sachant qu'il a $k$ jetons.

Alors
$$\begin{cases}
  \forall k \geqslant 1, q_{k} = p_{g} q_{k + 1} + \sum_{d = 1}^{n - 1} f_{d} p_{d} q_{k - d}
\\\forall k \in [\![0, -(n - 2)]\!], q_{k} = 0
\end{cases}$$

Calculons l'équation caractéristique associée à cette suite récurrente :

$$1 = p_{g} r + \sum_{d = 1}^{n - 1} f_{d} p_{d} r^{-d}$$

L'inconnue est $r$ et les solutions sont les $(r_i)_{i \in [\![0, n - 1]\!]}$ avec $r_0 = 1$. Nous notons aussi $\forall i \in [\![0, n - 1]\!], x_{i} = 1 / r_{i}$ où les $(x_i)_{i \in [\![0, n - 1]\!]}$ sont solutions du polynôme

$$P(X) = \sum_{d = 1}^{n - 1} f_{d} p_{d} X^{d + 1} - X + p_{g}$$

Nous remarquons que $P(X) = (X - 1)Q(X)$ avec

$$Q(X) = \sum_{l = 1}^{n - 1} c_{l} X^{l} - p_{g} \textrm{ où } \forall l \in [\![1, n - 1]\!], c_{l} = \sum_{d = l}^{n - 1} f_{d} p_{d}$$

Montrons que les racines de $Q$ sont de module strictement supérieur à $1$. De fait, supposons par l'absurde qu'il existe $x \in \mathbb{C}$ tel que $Q(x) = 0$ et $|x| \leqslant 1$. Alors 

$$\begin{align*}Q(x) = 0
  &\implies \sum_{l = 1}^{n - 1} c_{l} x^{l} - p_{g} = 0
\\&\implies \sum_{l = 1}^{n - 1} \sum_{d = l}^{n - 1} f_{d} p_{d} x^{l} + \sum_{d = 1}^{n - 1} f_{d} p_{d} = 1
\\&\implies \sum_{d = 1}^{n - 1} \sum_{l = 1}^{d} f_{d} p_{d} x^{l} + \sum_{d = 1}^{n - 1} f_{d} p_{d} = 1
\\&\implies \sum_{d = 1}^{n - 1} f_{d} p_{d} \sum_{l = 0}^{d} x^{l} = 1
\end{align*}$$

Or,

$$\begin{align*}1
  &= \left| \sum_{d = 1}^{n - 1} f_{d} p_{d} \sum_{l = 0}^{d} x^{l} \right|
\\&\leqslant \sum_{d = 1}^{n - 1} f_{d} p_{d} \sum_{l = 0}^{d} \left| x
^{l} \right|
\\&\leqslant \sum_{d = 1}^{n - 1} f_{d} p_{d} (d + 1)
\\&=\frac{n}{n + z}
\\&<1
\end{align*}$$

Absurde.

Donc, $\forall i \in [\![1, n - 1]\!], 1 < |x_{i}|$ et $\forall i \in [\![1, n - 1]\!], |r_{i}| < 1$.

Supposons les $(r_j)_{j \in [\![1, n - 1]\!]}$ distincts, alors $q_{k} = \sum_{i = 0}^{n - 1} \mu_{i} r_{i}^{k}$. Or, $\mathbb{E}(T_{m}) > 0$, donc $\lim_{k \to \infty} q_{k} = 1$. Et comme
$$\begin{align*}\lim_{k \to \infty} q_{k}
  &= \lim_{k \to \infty} \sum_{i = 0}^{n - 1} \mu_{i} r_{i}^{k}
\\&= \mu_{0} + \lim_{k \to \infty} \sum_{i = 1}^{n - 1} \mu_{i} r_{i}^{k}
\\&= \mu_{0}
\end{align*}$$

Nous déduisons $\mu_{0} = 1$ et donc $q_{k} = 1 + \sum_{i = 1}^{n - 1} \mu_{i} r_{i}^{k}$.

Nous déterminons les $n - 1$ autres coefficients $(\mu_{i})_{i \in [\![1, n - 1]\!]}$ grâce aux conditions initiales $\forall k \in [\![0, -(n - 2)]\!], q_{k} = 0$. Nous pouvons écrire ces conditions sous la forme matricielle suivante :

$$\begin{align*}\forall k \in [\![0, n - 2]\!], \sum_{i = 1}^{n - 1} \mu_{i} x_{i}^{k} = -1
  &\implies \begin{cases}
                  \mu_{1} x_{1}^{0} &+ &\cdots + \mu_{n - 1} x_{n - 1}^{0} &= -1
                \\\vdots &&&\vdots
                \\\mu_{1} x_{1}^{n - 2} &+ &\cdots + \mu_{n - 1} x_{n - 1}^{n - 2} &= -1
            \end{cases}
\\&\implies \begin{bmatrix}
                  x_{1}^{0} & \cdots & x_{n - 1}^{0}
                \\\vdots & & \vdots
                \\x_{1}^{n - 2} & \cdots & x_{n - 1}^{n - 2}
            \end{bmatrix}
            \begin{bmatrix}
                  \mu_{1}
                \\\vdots
                \\\mu_{n - 1}
            \end{bmatrix}
            =-\begin{bmatrix}
                  1
                \\\vdots
                \\1
            \end{bmatrix}
\\&\implies V \Mu = -1_{(n - 1) \times 1}
\\&\implies \Mu = -V^{-1}1_{(n - 1) \times 1}
\end{align*}$$

Où $V = (x_{j}^{i - 1})_{i, j \in [\![1, n - 1]\!]}$ désigne la matrice de Vandermonde associée aux $(x_{i})_{i \in [\![1, n - 1]\!]}$, inversible puisque nous avons supposé les racines simples, d'inverse $V^{-1}_{i} = L_{i}(X)$ polynôme interpolateur de Lagrange associé aux $(x_{i})_{i \in [\![1, n - 1]\!]}$. Ainsi,
$$\forall i \in [\![1, n - 1]\!], \mu_{i} = -L_{i}(1) = -\prod_{j \neq i} \frac{1 - x_{j}}{x_{i} - x_{j}}

Calculons $q_1$, la probabilité de survie en commençant avec un seul jeton.

$$\begin{align*}q_{1}
  &= 1 + \sum_{i = 1}^{n - 1} \mu_{i} r_{i}
\\&= 1 + \sum_{i = 1}^{n - 1} \frac{\mu_{i}}{x_{i}}
\\&= 1 + \frac{\sum_{i = 1}^{n - 1} \mu_{i} \prod_{j \neq i} x_{j}}{\prod x_{j}}
\\&= 1 + \frac{\sum_{i = 1}^{n - 1} \mu_{i} \prod_{j \neq i} x_{j}}{(-1)^{n - 1} Q(0)}
\end{align*}$$

Or

$$\begin{align*}\prod_{j \neq i} x_{j}
  &= \frac{(-1)^{n - 1}}{x_{i}} \left[ (-1)^{n - 1} \prod x_{j} \right]
\\&= \frac{(-1)^{n - 1}}{x_{i}} \left[ Q(0) - Q(x_{i}) \right]
\\&= \frac{(-1)^{n - 1}}{x_{i}} \left[ -\sum_{l = 1}^{n - 1} c_{l} x_{i}^{l} \right]
\\&= (-1)^{n} \sum_{l = 1}^{n - 1} c_{l} x_{i}^{l - 1}
\end{align*}$$

Donc

$$\begin{align*}q_{1}
  &= 1 + \frac{\sum_{i = 1}^{n - 1} \mu_{i} (-1)^{n} \sum_{l = 1}^{n - 1} c_{l} x_{i}^{l - 1}}{(-1)^{n - 1} Q(0)}
\\&= 1 - \frac{\sum_{l = 1}^{n - 1} c_{l} \sum_{i = 1}^{n - 1} \mu_{i} x_{i}^{l - 1}}{Q(0)}
\\&= 1 + \frac{\sum_{l = 1}^{n - 1} c_{l}}{Q(0)}
\\&= \frac{Q(1)}{Q(0)}
\\&= -\frac{\sum_{l = 1}^{n - 1} \sum_{d = l}^{n - 1} f_{d} p_{d} - 1 + \sum_{d = 1}^{n - 1} f_{d} p_{d}}{p_{g}}
\\&= \frac{1 - \sum_{d = 1}^{n - 1} (d + 1) f_{d} p_{d}}{p_{g}}
\\&= \frac{z}{p_{g}(n + z)}
\end{align*}$$
